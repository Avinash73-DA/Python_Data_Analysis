## US Accidents (2016 - 2023)

### This is a countrywide car accident dataset that covers 49 states of the USA. The accident data were collected from February 2016 to March 2023, using multiple APIs that provide streaming traffic incident (or event) data. These APIs broadcast traffic data captured by various entities, including the US and state departments of transportation, law enforcement agencies, traffic cameras, and traffic sensors within the road networks. The dataset currently contains approximately 7.7 million accident records.

```python
#START TIME - It is in the format of object data type 
df.Start_Time[0]
```

```python
#Converting the Time - object type into date type using this code
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')
```

```python
#Using the dt.hour function to get the hour in the each time data
hours = df['Start_Time'].dt.hour
hours
```

###  Accident Trends

```python
#We used the "norm_hist=True" toget the percentage of teh values on the y axis
hist_data = sns.distplot(hours,bins = 24,kde = False, norm_hist=True)
for patch, left_bin in zip(hist_data.patches, hist_data.patches):
    for patch in hist_data.patches:
        bin_center = patch.get_x() + patch.get_width() / 2  # Find the center of the bin
        if (6 <= bin_center < 9) or (14 <= bin_center < 19):
            patch.set_facecolor('darkblue')
        else:
            patch.set_facecolor('lightblue')
```

### Accidents on weekly basis as 7 days...

```python
day = df['Start_Time'].dt.dayofweek
day
```

```python
#During weekends no of accidents were lower as compared to the weekdays 

weeks = sns.distplot(day,bins = 7,kde = False, norm_hist=True)
patches = weeks.patches
for i, patch in enumerate(patches):
    if i < 5:
        patch.set_facecolor('skyblue')   # First five bins (0 to 4)
    elif i == 5:
        patch.set_facecolor('cornflowerblue')  # Sixth bin
    elif i == 6:
        patch.set_facecolor('dodgerblue')    # Seventh bin
```

### Accidents combined on saturday and sunday to compare it with monday

```python
sunSats = df.Start_Time[df.Start_Time.dt.dayofweek == 6 | (df.Start_Time.dt.dayofweek == 5)]
sunSats
```

```python
#Comparing the sunday and saturday distribution of the accidents it occurs as same as on the monday 

ss = sns.distplot(sunSats.dt.hour,bins = 24,kde = False, norm_hist=True)
for patch, left_bin in zip(hist_data.patches, ss.patches):
    for patch in ss.patches:
        bin_center = patch.get_x() + patch.get_width() / 2  # Find the center of the bin
        if (6 <= bin_center < 9) or (14 <= bin_center < 17):
            patch.set_facecolor('indigo')
        else:
            patch.set_facecolor('mediumpurple')
```

### Accidents monthly wise from  2016 - 2023

```python
month = df['Start_Time'].dt.month
month
```

```python
#The rate of accident is higher over the beginning and the ending of the year may be due to the festive season
mn = sns.distplot(month,bins = 12,kde = False, norm_hist=True)
for patch, left_bin in zip(hist_data.patches, mn.patches):
    for patch in mn.patches:
        bin_center = patch.get_x() + patch.get_width() / 2  # Find the center of the bin
        if (10 <= bin_center < 12) or (1 <= bin_center < 2):
            patch.set_facecolor('darkcyan')
        else:
            patch.set_facecolor('paleturquoise')
```

 ### Comparision of four years 

```python
year_of_4 = df.Start_Time[(df.Start_Time.dt.year == 2023) | (df.Start_Time.dt.year == 2022) | (df.Start_Time.dt.year == 2021) | (df.Start_Time.dt.year == 2021)]
year_of_4
```

```python
yr = sns.distplot(year_of_4.dt.month,bins = 12,kde = False, norm_hist=True)
for patch, left_bin in zip(hist_data.patches, yr.patches):
    for patch in yr.patches:
        bin_center = patch.get_x() + patch.get_width() / 2  # Find the center of the bin
        if (11 <= bin_center < 12) or (1 <= bin_center < 3):
            patch.set_facecolor('darkcyan')
        else:
            patch.set_facecolor('paleturquoise')
```

### 6,Peak accident timings:

    The number of accidents is lower on weekends compared to weekdays,likely due to office and school commutes causing higher weekday accident rates.

    The distribution of accidents on Sunday and Saturday is almost the same as on Monday.

### 9, Monthly accident trends from 2016 - 2023:

    The rate of accidents is higher at the beginning and end of the year,   possibly due to the festive season, especially during October, November,  December, January, and February.

### 10, Accident trends from 2021 - 2023:

    During these years, there is a clear increase in the trend of accidents in January and February, which is greater than in November and December.
    
    The number of accidents has been reduced during the summer, especially  in October, over the past four years.
